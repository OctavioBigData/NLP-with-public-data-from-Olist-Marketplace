# **Projeto** **Engenharia** **de** **Dados** 

image.png

O projeto de Engenharia de Dados da Squad Koalas foi desenvolvido inteiramente no intuito de ser o mais próximo da vida real possível.

Conceitualmente **Databricks** é uma solução em cloud e também considerado um canivete suíço baseado em Apache Spark, pois permite que engenheiros de dados, cientistas de dados, analistas de negócios e engenheiros de IA trabalhem de forma colaborativa em um único lugar com suporte para diversas linguagens como : Python, Scala, R, SQL, além também de bibliotecas e estruturas de aprendizado profundo como : Tensorflow, Pytorch e Scikit-learn.

No **Databricks** temos fluxos de trabalhos simplificados e um workspace interativo que permite toda essa interação, além também de ser parceria com as principais soluções em cloud como: Azure, GCP e AWS.
Diante disso optamos por um projeto feito nesta ferramenta de ponta a ponta.

# **Etapa 1 - Landing Zone**

Landing Zone é a nossa zona de recebimento do dados brutos, onde criamos uma pasta para isto.

Nele fizemos uma leitura dos dataframes e um pré processamento, transformando os dados em **parquet** para então passar para a próxima zona.

# **Etapa** **2** - **Processing** **Zone**

Já com os dados transformados em Parquet, nesta etapa trabalhamos todo o processo de ELT, normalização dos dados, análise exploratória, pré-processamento e machine learning.

Vale ressaltar que todos os notebooks utilizados para o desenvolvimento do projeto são os mesmo que constam no **github**, pois transformamos os códigos para **spark** utilizando as **bibliotecas databricks koalas** e **pyspark pandas**.
